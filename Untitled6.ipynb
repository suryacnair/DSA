```python
# Install required libraries (if not already installed)
!pip install transformers torch scikit-learn nltk matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, classification_report
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch
from torch.utils.data import Dataset
```

```python
# Initialize NLTK resources
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
# Load datasets
df_fake = pd.read_csv("/content/Fake (1).csv")
df_true = pd.read_csv("/content/True (1).csv")
```

```python
# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))
```

```python
# Preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])
    return text

# Clean and label datasets
df_true["text"] = df_true["text"].replace("(Reuters)", "", regex=True)
df_fake["target"] = 0
df_true["target"] = 1
df_fake = df_fake.drop(["title", "subject", "date"], axis=1)
df_true = df_true.drop(["title", "subject", "date"], axis=1)

# Combine datasets and shuffle
df = pd.concat([df_fake, df_true], axis=0).sample(frac=1).reset_index(drop=True)
df["text"] = df["text"].apply(preprocess_text)

# Reduce dataset size for faster testing (optional)
df = df.sample(frac=0.2).reset_index(drop=True)  # Use 20% of data for faster training
```

```python
# Visualizing word frequencies before and after preprocessing

from wordcloud import WordCloud

# Combine the text from the entire dataset
all_text = " ".join(df['text'].tolist())

# Create WordCloud before preprocessing
wordcloud_before = WordCloud(width=800, height=400, background_color="white").generate(all_text)

# Plot WordCloud for before preprocessing
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_before, interpolation="bilinear")
plt.title("WordCloud Before Preprocessing", fontsize=16)
plt.axis("off")
plt.show()

# Apply preprocessing
df['processed_text'] = df['text'].apply(preprocess_text)

# Combine the preprocessed text
all_text_processed = " ".join(df['processed_text'].tolist())

# Create WordCloud after preprocessing
wordcloud_after = WordCloud(width=800, height=400, background_color="white").generate(all_text_processed)

# Plot WordCloud for after preprocessing
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_after, interpolation="bilinear")
plt.title("WordCloud After Preprocessing", fontsize=16)
plt.axis("off")
plt.show()
```

```python
x = df["text"]
y = df["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state=42)
```

```python
# Vectorize text for traditional ML models
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)

# Train traditional ML models
lr = LogisticRegression()
lr.fit(xv_train, y_train)

dt = DecisionTreeClassifier(random_state=42)
dt.fit(xv_train, y_train)

gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(xv_train, y_train)

rfc = RandomForestClassifier(random_state=42)
rfc.fit(xv_train, y_train)

voting_clf = VotingClassifier(
    estimators=[('Logistic Regression', lr), ('Decision Tree', dt), ('Random Forest', rfc)],
    voting='hard'
)
voting_clf.fit(xv_train, y_train)
```

```python
# BERT dataset class
class NewsDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts.iloc[idx]
        label = self.labels.iloc[idx]
        encoding = self.tokenizer(
            text,
            max_length=self.max_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(),
            "attention_mask": encoding["attention_mask"].squeeze(),
            "labels": torch.tensor(label, dtype=torch.long)
        }

# Initialize tokenizer and BERT model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# Training arguments with optimizations
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,  # Use 1 epoch for quicker testing
    per_device_train_batch_size=8,  # Adjust batch size if necessary
    per_device_eval_batch_size=16,
    warmup_steps=100,
    weight_decay=0.01,
    logging_dir=None,  # Disable W&B logging
    logging_steps=10,
    evaluation_strategy="epoch",
    fp16=True,  # Enable mixed precision training
)

# Prepare datasets for BERT
train_dataset = NewsDataset(x_train, y_train, tokenizer, max_length=128)
test_dataset = NewsDataset(x_test, y_test, tokenizer, max_length=128)

# Train the BERT model
trainer = Trainer(
    model=bert_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()

# Evaluate BERT
bert_preds = trainer.predict(test_dataset).predictions.argmax(axis=1)
bert_accuracy = accuracy_score(y_test, bert_preds)
print("BERT Accuracy:", bert_accuracy)
```

```python
# Model Evaluation and Performance Visualization

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Evaluate all models (including BERT)
def evaluate_model(model, name, plot_cm=False):
    predictions = model.predict(xv_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{name} Accuracy: {accuracy}")
    print(f"{name} Classification Report:\n{classification_report(y_test, predictions)}")

    # Plot confusion matrix
    if plot_cm:
        cm = confusion_matrix(y_test, predictions)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fake", "True"])
        disp.plot(cmap='Blues', values_format='d')
        plt.title(f"{name} Confusion Matrix")
        plt.show()
```

```python
# Evaluate and plot for all traditional models
evaluate_model(lr, "Logistic Regression", plot_cm=True)
evaluate_model(dt, "Decision Tree", plot_cm=True)
evaluate_model(gbc, "Gradient Boosting", plot_cm=True)
evaluate_model(rfc, "Random Forest", plot_cm=True)
evaluate_model(voting_clf, "Voting Classifier", plot_cm=True)

# For BERT evaluation
bert_preds = trainer.predict(test_dataset).predictions.argmax(axis=1)
bert_accuracy = accuracy_score(y_test, bert_preds)
print("BERT Accuracy:", bert_accuracy)
```

```python
# Compare Model Performance
model_names = [
    "Logistic Regression",
    "Decision Tree",
    "Gradient Boosting",
    "Random Forest",
    "Voting Classifier",
    "BERT"
]

# Collect accuracy scores for each model
model_accuracies = [
    accuracy_score(y_test, lr.predict(xv_test)),
    accuracy_score(y_test, dt.predict(xv_test)),
    accuracy_score(y_test, gbc.predict(xv_test)),
    accuracy_score(y_test, rfc.predict(xv_test)),
    accuracy_score(y_test, voting_clf.predict(xv_test)),
    bert_accuracy
]

# Plotting accuracy comparison
plt.figure(figsize=(10, 6))
plt.bar(model_names, model_accuracies, color=['#FFCDB2', '#FFB4A2', '#E5989B', '#B5838D', '#6D6875', '#C5C6D0'])
plt.title('Model Performance Comparison', fontsize=16)
plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.ylim(0, 1)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

from sklearn.metrics import precision_score, recall_score, f1_score

# Precision, Recall, F1 Score Comparison
precisions = [
    precision_score(y_test, lr.predict(xv_test)),
    precision_score(y_test, dt.predict(xv_test)),
    precision_score(y_test, gbc.predict(xv_test)),
    precision_score(y_test, rfc.predict(xv_test)),
    precision_score(y_test, voting_clf.predict(xv_test)),
    precision_score(y_test, bert_preds)
]

recalls = [
    recall_score(y_test, lr.predict(xv_test)),
    recall_score(y_test, dt.predict(xv_test)),
    recall_score(y_test, gbc.predict(xv_test)),
    recall_score(y_test, rfc.predict(xv_test)),
    recall_score(y_test, voting_clf.predict(xv_test)),
    recall_score(y_test, bert_preds)
]

f1_scores = [
    f1_score(y_test, lr.predict(xv_test)),
    f1_score(y_test, dt.predict(xv_test)),
    f1_score(y_test, gbc.predict(xv_test)),
    f1_score(y_test, rfc.predict(xv_test)),
    f1_score(y_test, voting_clf.predict(xv_test)),
    f1_score(y_test, bert_preds)
]

# Plot Precision, Recall, F1 Score comparison
metrics_df = pd.DataFrame({
    'Model': model_names,
    'Precision': precisions,
    'Recall': recalls,
    'F1 Score': f1_scores
})

metrics_df.set_index('Model').plot(kind='bar', figsize=(12, 6))
plt.title('Model Performance Comparison (Precision, Recall, F1)', fontsize=16)
plt.ylabel('Score', fontsize=14)
plt.xlabel('Models', fontsize=14)
plt.xticks(rotation=45)
plt.show()

# Plot learning curves for BERT
# This assumes you have `trainer` with `logging_dir` set and logs are being saved.

# Visualizing BERT training and validation loss
import matplotlib.pyplot as plt

history = trainer.state.log_history

# Extract loss values from the logs
train_loss = [log['loss'] for log in history if 'loss' in log]
eval_loss = [log['eval_loss'] for log in history if 'eval_loss' in log]

# Plot training and evaluation loss curves
plt.figure(figsize=(10, 5))
plt.plot(train_loss, label="Training Loss")
plt.plot(eval_loss, label="Evaluation Loss")
plt.title("BERT Training and Evaluation Loss", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend()
plt.show()
```

```python
# Manual testing function
def manual_testing(news):
    sample_data = {"text": [news]}
    sample_df = pd.DataFrame(sample_data)
    sample_df["text"] = sample_df["text"].apply(preprocess_text)

    # Predict with traditional models
    sample_vectorized = vectorizer.transform(sample_df["text"])
    print("Predictions for the input news article:")
    print("Logistic Regression:", "True" if lr.predict(sample_vectorized)[0] else "Fake")
    print("Decision Tree:", "True" if dt.predict(sample_vectorized)[0] else "Fake")
    print("Gradient Boosting:", "True" if gbc.predict(sample_vectorized)[0] else "Fake")
    print("Random Forest:", "True" if rfc.predict(sample_vectorized)[0] else "Fake")
    print("Voting Classifier:", "True" if voting_clf.predict(sample_vectorized)[0] else "Fake")

    # Move the model and inputs to GPU
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    bert_model.to(device)

    # Predict with BERT
    bert_encoding = tokenizer(
        news,
        max_length=128,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    input_ids = bert_encoding["input_ids"].to(device)
    attention_mask = bert_encoding["attention_mask"].to(device)

    # Perform inference on GPU
    bert_output = bert_model(input_ids, attention_mask=attention_mask)
    bert_pred = torch.argmax(bert_output.logits, dim=1).item()
    print("BERT:", "True" if bert_pred == 1 else "Fake")
```

```python
# Test manual input
news_article = input("Enter a news article to classify as Fake or True: ")
manual_testing(news_article)
```
